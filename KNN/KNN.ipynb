```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")
```

```python
dfx = pd.read_csv("Datasets/xdata.csv")
dfy = pd.read_csv("Datasets/ydata.csv")
```

```python
X = dfx.values
Y = dfy.values

X = X[:,1:]
Y = Y[:,1:].reshape((-1,))

#print(X)

print(X.shape)
print(Y.shape)

#print(Y)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
query_x = np.array([2,3])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query_x[0],query_x[1],color="red")
plt.show()
```

```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
def knn(X,Y,queryPoint,K=5):
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = distance(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    vals = sorted(vals)
    # Pick the Nearest/First K points
    
    vals = vals[:K]
    vals = np.array(vals)
    
    new_vals = np.unique(vals[:,1], return_counts=True)
    #print(new_vals)
    
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    #print(index)
    #print(pred)
    #print(vals)
    #return vals
    return pred
```

```python
pred = knn(X,Y,query_x)
print(pred)
```
