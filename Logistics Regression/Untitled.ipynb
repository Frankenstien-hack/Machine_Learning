```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("seaborn")
```

```python
mean1 = np.array([1,0.5])
covar1 = np.array([[1,0.1],[0.1,1.2]])

mean2 = np.array([4,5])
covar2 = np.array([[1.21,0.1],[0.1,1.3]])

dist1 = np.random.multivariate_normal(mean1,covar1,500)
dist2 = np.random.multivariate_normal(mean2,covar2,500)

print(dist1.shape)
print(dist2.shape)
```

```python
plt.figure(0)

plt.scatter(dist1[:,0],dist1[:,1],label="CLASS 0")
plt.scatter(dist2[:,0],dist2[:,1],label="CLASS 1",color="r",marker="^")

plt.xlim(-5,10)
plt.ylim(-5,10)
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```
