```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv("Datasets/linearX.csv")
dfy = pd.read_csv("Datasets/linearY.csv")

x = dfx.values
y = dfy.values

print(x.shape)
print(y.shape)
```

```python
#Visualization and Preprocessing

plt.scatter(x,y)


X = (x-x.mean())/x.std()
Y = y

plt.scatter(X,Y)
plt.show()
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
def hypothesis(x,theta):
    
    return theta[0] + theta[1]*x

def error(X,Y,theta):
    
    m = X.shape[0]
    error = 0
    
    for i in range(m):
        hx = hypothesis(X[i],theta)
        error += (hx-Y[i])**2
        
    return error

def gradient(X,Y,theta):
    
    m = X.shape[0]
    grad = np.zeros((2,))
    
    for i in range(m):
        hx = hypothesis(X[i],theta)
        grad[0] += hx-Y[i]
        grad[1] += (hx-Y[i])*X[i]
        
    return grad

def gradientDescent(X,Y,learning_rate = 0.001):
    
    theta = np.zeros((2,))
    e = 1
    count = 0
    while(e > 0.001):
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        print(e)
        theta[0] = theta[0] - learning_rate * grad[0]
        theta[1] = theta[1] - learning_rate * grad[1]
        count += 1
    return theta,e,count
```
