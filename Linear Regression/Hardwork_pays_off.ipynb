```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv("Datasets/Walkatime dataset/Training Data/Linear_X_Train.csv")
dfy = pd.read_csv("Datasets/Walkatime dataset/Training Data/Linear_Y_Train.csv")

x = dfx.values
y = dfy.values

print(x.shape)
print(y.shape)
```

```python
plt.scatter(x,y)

X = (x-x.mean())/x.std()
Y = y

plt.scatter(X,Y)
plt.show()
#print(X.shape,Y.shape)
```

```python
def hypothesis(x,theta):

    return theta[0] + theta[1]*x

def error(X,Y,theta):

    m = X.shape[0]
    error = 0

    for i in range(m):
        hx = hypothesis(X[i],theta)
        error += (hx-Y[i])**2

    return error

def gradient(X,Y,theta):

    m = X.shape[0]
    grad = np.zeros((2,))

    for i in range(m):
        hx = hypothesis(X[i],theta)
        grad[0] += (hx-Y[i])
        grad[1] += (hx-Y[i])*X[i]

    return grad

def gradient_descent(X,Y,learning_rate = 0.00052):

    theta = np.zeros((2,))
    itr = 0
    itr_max = 100
    error_list = []
    
    while(itr<=itr_max):    
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]

        itr += 1

    return theta,error_list
```

```python
finaltheta,error = gradient_descent(X,Y)
print(finaltheta)
```

```python
plt.plot(error)
plt.show
```

```python
x_test = pd.read_csv("Datasets/Walkatime dataset/Test Cases/Linear_X_Test.csv")
# X_test = x_test.values
plt.scatter(X,Y,label="Training data")
plt.plot(x_test,hypothesis(x_test,finaltheta),label="Testing data",color="orange")
plt.legend()
plt.show()
```
