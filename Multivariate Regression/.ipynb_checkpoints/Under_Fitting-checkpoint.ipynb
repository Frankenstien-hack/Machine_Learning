```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dfx = pd.read_csv("Datasets/Locally Weighted Regression/weightedX.csv")
dfy = pd.read_csv("Datasets/Locally Weighted Regression/weightedY.csv")

x = dfx.values
y = dfy.values

print(x.shape)
print(y.shape)
```

```python
#plt.scatter(x,y)
X = (x-x.mean())/x.std()
Y = y

plt.scatter(X,Y)
plt.show()
```

```python
model = LinearRegression()
model.fit(X,Y)
output = model.predict(X)
```

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

```python
model.score(X,Y)
```

```python
plt.scatter(X,Y,label = "Data")
plt.plot(X,output,color = "orange",label = "Prediction")
plt.legend()
plt.show()
```
